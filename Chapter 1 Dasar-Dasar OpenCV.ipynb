### 2. Pengenalan OpenCV ###
OpenCV adalah sebuah library yang dibangun dari bahasa pemrograman C dan digunakan untuk pengolahan citra maupun video. Library ini telah banyak dikompilasi untuk bahasa pemrograman lain seperti C++, Python, dan Java. 

### 3. Instalasi OpenCV ###
Dalam tutorial ini, kita akan belajar untuk setup OpenCV-Python di Ubuntu. Di sini, kita menggunakan OpenCV yang sudah mendukung bahasa Python.  

OpenCV-Python dapat diinstal di Ubuntu dengan dua cara:
  1. Instal dari pre-built file binary yang berasal dari repo Ubuntu  
  1. Kompilasi dari source asli.

Hal penting lainnya adalah instalasi library tambahan, yaitu **numpy** dan **matplotlib**.  
Di sini, kita akan mencontohkan instalasi di Ubuntu dan Windows.  

Instalasi di Ubuntu:
1. Buka terminal, kemudian jalankan perintah berikut:  
<code>$ sudo apt-get install python3-opencv</code>  

2. Tunggu hingga instalasi selesai. Jika sudah, ketik perintah:  
<code>$ python3</code>  
maka, Anda akan melihat console python.  
3. Untuk mengecek apakah hasil instalasi sudah berhasil, jalan perintah berikut pada console Python:  

```
import cv2 as cv  
print(cv.__version__)
```

4. Berikut adalah hasil output dari versi OpenCV yang diinstal  
<code>4.4.0</code>  

Instalasi di Windows:  
1. Silakan mengunduh package python dari [sini](https://www.python.org/downloads/). Pilih yang Python versi 3.  
1. Instalasi numpy dan matplotlib  

    <code>pip3 install numpy</code>  
    <code>pip3 install matplotlib</code>  
    
1. Jika sudah selesai instalasi, buka Python IDLE dari command prompt. Ketik `import numpy` dan jalankan untuk memastika bahwa **numpy** sudah diinstal.
1. Unduh OpenCV yang terbaru dari [sini](https://github.com/opencv/opencv/releases)
1. Buka folder **opencv/build/python/2.7** dan copy file **cv2.pyd** ke folder **C:/Python3/lib/site-packages**
1. Buka Python IDLE dan jalan perintah berikut untuk memastikan OpenCV telah berhasil  

```
import cv2 as cv  
print(cv.__version__)
```  


### 4. Membaca atau Load Citra ###  
Silakan membuat sebuah file dengan nama **main.py**. Isikan kode berikut ke dalam file **main.py**. Buat sebuah folder, simpan file ini di dalam folder yang baru saja dibuat. Silakan cari citra apapun dengan ekstensi .png atau .jpg. Taruh citra tersebut di dalam folder bersamaan dengan file **main.py**

```
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

img = cv.imread('nama_citra.jpg')
```

### 5. Operasi pada Citra ###
Di sini, kita akan mengubah citra dari RGB ke Grayscale. Perintahnya adalah

```
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

img = cv.imread('nama_citra.jpg')
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

```

### 6. Menampilkan Citra ###  

```
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

img = cv.imread('nama_citra.jpg')
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

plt.figure(figsize=(20, 20))
plt.subplot(1,2,1)
plt.imshow(img,cmap="gray")
plt.subplot(1,2,2)
plt.imshow(img_gray,cmap="gray")
plt.show()

```
