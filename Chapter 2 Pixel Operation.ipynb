```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

**A. Image Addition**  
Penjumlahan citra sama seperti penjumlahan pada matriks. Jika $\mathbf{A}$ adalah sebuah matriks berukuran n x m, dan $b$ adalah skalar, dan $\mathbf{C}$ adalah hasil penjumlahannya, maka  
$\mathbf{C} = \mathbf{A} + b$  
Contoh:  
$\mathbf{A} = \begin{bmatrix}1 \\ 2 \\ 3\end{bmatrix}$  dan $b = 2$  
maka $\mathbf{C} = \begin{bmatrix}3 \\ 4 \\ 5 \end{bmatrix}$  
  
Berikut ini adalah contoh penjumlahan pada citra  


```python
lego_a = cv2.imread('dataset/scene1.png',1)
lego_a = cv2.cvtColor(lego_a, cv2.COLOR_BGR2RGB)

lego_b = cv2.imread('dataset/scene2.png',1)
lego_b = cv2.cvtColor(lego_b, cv2.COLOR_BGR2RGB)

lego_add = cv2.add(lego_a, lego_b)
plt.figure(figsize=(20, 20))
plt.subplot(1,3,1)
plt.imshow(lego_a)
plt.subplot(1,3,2)
plt.imshow(lego_b)
plt.subplot(1,3,3)
plt.imshow(lego_add)
plt.show()
```

**B. Image Subtraction**  
Pengurangan citra sama seperti pengurangan pada matriks. Jika $\mathbf{A}$ adalah sebuah matriks berukuran n x m, dan $b$ adalah skalar, dan $\mathbf{C}$ adalah hasil penjumlahannya, maka  
$\mathbf{C} = \mathbf{A} - b$  
Contoh:  
$\mathbf{A} = \begin{bmatrix}3 \\ 4 \\ 2\end{bmatrix}$  dan $b = 2$  
maka $\mathbf{C} = \begin{bmatrix}1 \\ 2 \\ 0 \end{bmatrix}$  
  
Berikut ini adalah contoh pengurangan pada citra  


```python
lego_subtract = cv2.subtract(lego_a, lego_b)
plt.figure(figsize=(20, 20))
plt.subplot(1,3,1)
plt.imshow(lego_a)
plt.subplot(1,3,2)
plt.imshow(lego_b)
plt.subplot(1,3,3)
plt.imshow(lego_subtract)
plt.show()
```

**C. Image Blending**  
Pada poin A, fungsi cv2.add adalah seperti berikut:  
cv2.add(200, 70) = 270 = 255  
sedangkan operator '+' pada $200 + 70$ adalah modulus, sehingga  
200 % 70 = 60  
  
Ada teknik lain yang disebut sebagai image blending. Meskipun sama-sama penjumlahan citra, tapi ada perbedaan pada sisi formulanya.  
Jika $f(x)$ adalah citra pertama, dan $g(x)$ adalah citra kedua, maka blending dari citra pertama dan kedua maka  
  
$h(x) = \alpha*f(x) + \beta*g(x) + \gamma$  
  
di mana $\alpha$ dan $\beta$ adalah bobot pada masing-masing element citra, sedangkan $\gamma$ adalah suatu nilai skalar
    
Berikut ini adalah contoh blending pada citra  

```python
lego_blending = cv2.addWeighted(lego_a, 0.7, lego_b, 0.5, 0)
plt.figure(figsize=(20, 20))
plt.subplot(1,3,1)
plt.imshow(lego_a)
plt.subplot(1,3,2)
plt.imshow(lego_b)
plt.subplot(1,3,3)
plt.imshow(lego_blending)
plt.show()
```
