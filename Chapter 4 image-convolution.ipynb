## A. Image Blurring ##

```python
import cv2 as cv
import numpy as np

import matplotlib.pyplot as plt

img = cv.imread("dataset/cameraman.png",1)
kernel = np.ones((3, 3), np.float32) / 9

img_out = cv.filter2D(img, -1, kernel)
plt.figure(figsize=(20, 20))
plt.subplot(1,2,1)
plt.imshow(img,cmap="gray")
plt.subplot(1,2,2)
plt.imshow(img_out,cmap="gray")
plt.show()
```

## B. Image Sharpening ##

```python
import cv2 as cv
import numpy as np

import matplotlib.pyplot as plt

img = cv.imread("dataset/cameraman.png",1)
kernel = np.array([
    [0, -1, 0],
    [-1, 5,-1],
    [0, -1, 0]
])


img_out = cv.filter2D(img, -1, kernel)
plt.figure(figsize=(20, 20))
plt.subplot(1,2,1)
plt.imshow(img,cmap="gray")
plt.subplot(1,2,2)
plt.imshow(img_out,cmap="gray")
plt.show()
```

```python
import cv2 as cv
import numpy as np

import matplotlib.pyplot as plt

img = cv.imread("dataset/cameraman.png",1)
kernel = np.array([
    [1, 0, -1],
    [2, 0, -2],
    [1, 0, -1]
])

img_out = cv.filter2D(img, -1, kernel)
plt.figure(figsize=(20, 20))
plt.subplot(1,2,1)
plt.imshow(img,cmap="gray")
plt.subplot(1,2,2)
plt.imshow(img_out,cmap="gray")
plt.show()
```
