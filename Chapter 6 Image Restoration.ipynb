```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

```

```python
def noise(img,snr):
    h=img.shape[0]
    w=img.shape[1]
    img1=img.copy()
    sp=h*w # Calculate the number of image pixels
    NP=int(sp*(1-snr)) # Calculate the number of image salt and pepper noise points
    for i in range (NP):
        randx=np.random.randint(1,h-1) # Generate a random integer between 1 and h-1
        randy=np.random.randint(1,w-1) # Generate a random integer between 1 and w-1
        
        if np.random.random()<=0.5: # np.random.random() generates a floating point number between 0 and 1.
            img1[randx,randy]=0
        else:
            img1[randx,randy]=255
    return img1
```

```python
img = cv.imread("dataset/cameraman.png",1)

img_out = noise(img, 0.6)
plt.figure(figsize=(20, 20))
plt.subplot(1,2,1)
plt.imshow(img,cmap="gray")
plt.subplot(1,2,2)
plt.imshow(img_out,cmap="gray")
plt.show()
```

```python
import cv2
import numpy as np

img = cv.imread("dataset/cameraman.png",1)

img_noised = noise(img, 0.6)
median = cv2.medianBlur(img_noised, 5)

plt.figure(figsize=(20, 20))
plt.subplot(1,2,1)
plt.imshow(img_noised,cmap="gray")
plt.subplot(1,2,2)
plt.imshow(median,cmap="gray")
plt.show()
```

```python
import cv2
import numpy as np

img = cv.imread("dataset/cameraman.png",1)

img_noised = noise(img, 0.6)
img_gauss = cv2.GaussianBlur(img_noised, (5,5),0)

plt.figure(figsize=(20, 20))
plt.subplot(1,2,1)
plt.imshow(img_noised,cmap="gray")
plt.subplot(1,2,2)
plt.imshow(img_gauss,cmap="gray")
plt.show()
```
